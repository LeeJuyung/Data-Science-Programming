### 반복문을 통하여 전체 행의 정보 가져오기
### - nested loop (이중 반복문)을 응용하여 리스트로 전체 테이블의 행을 출력하기

```python
table_data = []

# Nested loop를 사용하여 전체 10개 행의 정보를 가져옴
for row in rows:
    columns = row.find_all('td')  # 현재 행의 모든 열을 찾음
    row_data = []  # 현재 행의 열 값을 저장할 리스트 초기화
    for column in columns:
        row_data.append(column.get_text())  # 현재 열의 값을 리스트에 추가
    table_data.append(row_data)  # 현재 행의 정보를 리스트에 추가
    if len(table_data) == 10:  # 10개 행을 가져왔을 때 루프 종료
        break

# 결과 출력
for row in table_data:
    print(row)
```

# In-class exercise - url을 수정하여 10페이지까지의 데이터 수집

```python
import requests
from bs4 import BeautifulSoup

# 결과를 저장할 리스트 초기화
table_data = []

# 1부터 10까지 페이지를 순환하면서 데이터 가져오기
for page in range(1, 11):
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page}&sido=&gugun=&store='
    
    # 웹 페이지 요청
    response = requests.get(url)
    
    # BeautifulSoup을 사용하여 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 테이블의 각 행을 찾음
    rows = soup.find_all('tr')
    
    # 각 행의 데이터를 가져와서 리스트에 추가
    for row in rows:
        columns = row.find_all('td')
        row_data = [column.get_text() for column in columns]
        if row_data:  # 빈 리스트가 아닌 경우에만 추가
            table_data.append(row_data)
    
    # 연결 종료
    response.close()

# 결과 출력
for row in table_data:
    print(row)
```

## In-class Exercise 1 - csv파일 저장 시 예외처리
- 매장서비스가 공백으로 출력되니 매장서비스 열은 제외하고 출력해 보자
- 이번에는 파일 저장 시 지역, 매장명, 현황, 주소, 전화번호의 열 이름 (header)를 넣어서 출력해 보자

```python
import csv
import requests
from bs4 import BeautifulSoup

# 결과를 저장할 리스트 초기화
table_data = []

# 페이지 1부터 10까지 데이터 가져오기
for page in range(1, 11):
    url = f'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo={page}&sido=&gugun=&store='
    
    # 웹 페이지 요청
    response = requests.get(url)
    
    # BeautifulSoup를 사용하여 HTML 파싱
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 테이블에서 각 행 찾기
    rows = soup.find_all('tr')
    
    # 각 행의 데이터 가져와서 딕셔너리에 추가하기
    for row in rows:
        columns = row.find_all('td')
        row_data = [column.get_text() for column in columns]
            
        # "Store service" 열을 제외한 데이터만 포함하는지 확인
        if len(row_data) == 6:
            data_dict = {
                "지역": row_data[0],
                "매장명": row_data[1],
                "현황": row_data[2],
                "주소": row_data[3],
                "전화번호": row_data[5]
            }
            table_data.append(data_dict)

    # 연결 종료
    response.close()

# 열 이름 (헤더)
header = ["지역", "매장명", "현황", "주소", "전화번호"]

# 결과를 저장할 CSV 파일 경로
csv_file_path = 'C:/Users/jylpi/OneDrive/Desktop/StoreInfo.csv'

# 데이터를 CSV 파일로 저장
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=header)
    
    # 헤더를 파일에 쓰기
    csv_writer.writeheader()
    
    # 데이터를 파일에 쓰기
    csv_writer.writerows(table_data)

print(f'Data has been saved to {csv_file_path}')
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.hollys.co.kr/store/korea/korStore2.do?pageNo=1&sido=&gugun=&store='
r = requests.get(url)

soup = BeautifulSoup(r.text, "html.parser")

table = soup.find('table')
rows = table.find('tbody').find_all('tr')

# 각 행을 순회하며 모든 정보 출력
for row in rows:
    cols = row.find_all('td')
    row_data = [col.get_text(strip=True) for col in cols]

    # col[4]에서 <img> 태그 정보를 가져오기 (있을 경우)
    img = cols[4].find('img')
    if img:
        img_info = img.get('alt')
        
        # "Store Services" 정보를 5번째 항목으로 대체
        row_data[4] = img_info
    
    print(row_data)
```
